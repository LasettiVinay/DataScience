```python
# imports
import pandas as pd
import numpy as np
import os
```

```python
print(os.path.pardir)
```

### import data

```python
# set the path of the raw data (from par dirctory)
rawDataPath = os.path.join(os.path.pardir, 'data', 'raw')
testFilePath = os.path.join(rawDataPath, 'test.csv')
trainFilePath = os.path.join(rawDataPath, 'train.csv')
```

```python
# get test and train dataframes
train_df = pd.read_csv(trainFilePath, index_col='PassengerId')
test_df = pd.read_csv(testFilePath, index_col='PassengerId')
```

```python
# check data type 
type(train_df)
```

### Basic Structure

```python
# use .info() to get the basic information about the data frame
train_df.info()
```

```python
test_df.info()
```

```python
test_df['Survived'] = -888 # Adding Survived with a default value
```

```python
df = pd.concat((train_df, test_df), axis=0)
```

```python
# filter rows based on the condition
male_passengers = df.loc[df.Sex == 'male']
print(male_passengers)
```

```python
print('Number of male passengers : {}'.format(len(male_passengers)))
```

```python
male_firstClassPassengers = df.loc[(df.Sex == 'male') & (df.Pclass == 1)]
print('Number of male passengers in 1st Class: {}'.format(len(male_firstClassPassengers)))
```

```python
male_2ndClassPassengers = df.loc[(df.Sex == 'male') & (df.Pclass == 2)]
male_3rdClassPassengers = df.loc[(df.Sex == 'male') & (df.Pclass == 3)]
print('Number of male passengers in 2nd Class: {}'.format(len(male_2ndClassPassengers)))
print('Number of male passengers in 3nd Class: {}'.format(len(male_3rdClassPassengers)))
```

### Summary Statistics

```python
df.describe()
```

#### numerical feature

```python

# centrality measure
print('Mean Fare  : {}'.format(df.Fare.mean()))
print('Median Fare  : {}'.format(df.Fare.median()))
```

```python
# disperssion measure
print('Min Fare  : {}'.format(df.Fare.min()))
print('Max Fare  : {}'.format(df.Fare.max()))
print('Fare range  : {}'.format(df.Fare.max()-df.Fare.min()))
print('Fare 25 Percentile  : {}'.format(df.Fare.quantile(.25)))
print('Fare 50 Percentile  : {}'.format(df.Fare.quantile(.50)))
print('Fare 75 Percentile  : {}'.format(df.Fare.quantile(.75)))
print('Variance Fare  : {}'.format(df.Fare.var()))
print('Standard Deviation Fare  : {}'.format(df.Fare.std()))
```
